```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('UNdata_Export_20180518_223029078.csv')
```

```python
usa = df[3824:3841]
aus = df[7:33]
fran = df[796:822]
germ = df[961:986]
isr = df[1354:1380]





```

```python
plt.figure(figsize=(15, 5))
plt.plot(usa['Year'], usa['Quantity'], color='green')
plt.title('US Solar Energy Consumption')
plt.ylabel('Terajoules')
plt.show()

# The following is a series of countries and their use of solar energy. They measure the "direct use of solar 
# thermal heat." The use of that energy has been measured in terajoules and the amounts shown are the total
# final energy consumption, correlating to a plot for each year shown at the bottom x-axis of the graphs.

# This is by no means an in depth analysis of green energy initiatives, but can indicate each country's initiatives
# and show possible difficulties and hinderances. Data was provided by the United Nations.

# This first graph plots the solar energy consumption of the United States. We can see it begins by trending downward
# for several years. It would be interesting to cross check the political environment of of the US during that time
# as well as how the military budget changed during several large events, including the war on terrorism.
```

```python
plt.figure(figsize=(15, 5))
plt.plot(aus['Year'], aus["Quantity"], color='red')
plt.title('Australia Solar Energy Consumption')
plt.ylabel('Terajoules')
plt.show()

# The next graph shows Australia. They also show a dip in solar energy use when US consumption was down at well
# followed by a rebound and rapid climb in the years to follow.
```

```python
plt.figure(figsize=(15, 5))
plt.plot(fran['Year'], fran['Quantity'], color='yellow')
plt.title('France Solar Energy Consumption')
plt.ylabel('Terajoules')
plt.show()

# The third graph demonstrates France's solar energy use and they have similar trends as with the previously shown
# countries. This can indicate several things including world economy health and political initiatives toward
# renewable energy sources. As we know green initiatives have been discussed and agreed upon by many countries
# and we can see this from the consumption levels rising quite considerably from around 2009.
```

```python
plt.figure(figsize=(15, 5))
plt.plot(germ['Year'], germ["Quantity"], color='blue')
plt.title('Germany Solar Energy Consumption')
plt.ylabel('Terajoules')
plt.show()

# At first glance it may seem the next graph showing Germany demonstrates a similar story. Upon further examination
# we can see it is quite different in that its green initiatives seem to be a bit ahead of others in that it has
# been steadily developing and implemented since the early 90's. Also note the consumption goes above 25,000
# terajoules while France's was peaking a mere 4,000. France's total population is somewhere in the vicinity of 75%
# that of Germany's. From this we can ascertain a large difference between initiatives and development.
```

```python
plt.figure(figsize=(15, 5))
plt.plot(isr['Year'], isr['Quantity'], color="orange")
plt.title('Israel Solar Energy Consumption')
plt.ylabel('Terajoules')
plt.show()

# This final graph was included to show a contrasting picture. First, note how high the points on the graph climb,
# 45,000 terajoules, very impressive for a small piece of land, the smallest country listed in my report.
# Between 2013-2014 there was a dramatic dip in solar use. A lot happened during this time as this land has been
# a hot bed for religious disputes and political turmoil. I'm not going to assert the exact cause for the drop,
# but it serves as a good contrasting demonstration that these initiatives can vary by country and can be heavily
# influenced not only by macro environmental factors but also by highly localized factors.
```
